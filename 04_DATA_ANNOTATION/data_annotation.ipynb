```python
import pandas as pd
from sklearn.datasets import load_iris

# Memuat dataset Iris
iris = load_iris()

# Membuat DataFrame dari fitur Iris
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Menambahkan label asli (ground truth) ke dalam DataFrame
df['species_asli'] = pd.Categorical.from_codes(
    iris.target, iris.target_names
)

# Fungsi pelabelan berdasarkan SOP
def label_sesuai_sop(row):
    petal_length = row['petal length (cm)']
    petal_width = row['petal width (cm)']

    if petal_length < 2.5 and petal_width < 0.8:
        return 'setosa'
    elif petal_length > 5.1 and petal_width > 1.8:
        return 'virginica'
    elif 3.0 <= petal_length <= 5.1:
        return 'versicolor'
    elif petal_length > 4.5:
        return 'virginica'
    else:
        return 'versicolor'

# Menerapkan SOP ke seluruh data
df['sop_label'] = df.apply(label_sesuai_sop, axis=1)

# Menampilkan perbandingan label
print("Perbandingan Label Asli dan Label Berdasarkan SOP:")
print(df[['species_asli', 'sop_label']].head(10))

print("\nContoh data dengan perbedaan label:")
print(df[df['species_asli'] != df['sop_label']].head())
```
