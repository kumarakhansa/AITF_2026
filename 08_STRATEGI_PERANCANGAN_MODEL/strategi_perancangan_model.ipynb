```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
cancer = load_breast_cancer()

df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df['diagnosis'] = cancer.target
```

```python
# Distribusi jumlah kelas
print("Distribusi kelas diagnosis:")
print(df['diagnosis'].value_counts())

# Distribusi proporsi kelas
print("\nProporsi kelas diagnosis:")
print(df['diagnosis'].value_counts(normalize=True))

# Visualisasi distribusi kelas
plt.figure(figsize=(6, 4))
sns.countplot(x='diagnosis', data=df)
plt.title("Distribusi Kelas Diagnosis")
plt.xticks([0, 1], ['Malignant', 'Benign'])
plt.show()
```

```python
corr_matrix = df.drop(columns='diagnosis').corr()

plt.figure(figsize=(16, 14))
sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    center=0,
    linewidths=0.3
)
plt.title("Heatmap Korelasi Antar Seluruh Fitur")
plt.show()

high_corr = corr_matrix.abs()
high_corr_pairs = high_corr.where(
    np.triu(np.ones(high_corr.shape), k=1).astype(bool)
)

high_corr_pairs = high_corr_pairs.stack().sort_values(ascending=False)

print("\nPasangan fitur dengan korelasi tertinggi:")
print(high_corr_pairs.head(10))
```

```python
target_corr = df.corr()['diagnosis'].sort_values(ascending=False)
print("\nKorelasi fitur terhadap target diagnosis:")
print(target_corr)
```

```python
# Pemisahan Fitur dan Target
X = df.drop(columns='diagnosis')
y = df['diagnosis']
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', DecisionTreeClassifier(random_state=42))
])
```

```python
scoring = ['accuracy', 'precision', 'recall', 'f1']

cv_results = cross_validate(
    pipeline,
    X,
    y,
    cv=5,
    scoring=scoring
)
```

```python
# Rata-rata hasil evaluasi
results = {
    'Accuracy': cv_results['test_accuracy'].mean(),
    'Precision': cv_results['test_precision'].mean(),
    'Recall': cv_results['test_recall'].mean(),
    'F1-Score': cv_results['test_f1'].mean()
}

results_df = pd.DataFrame(results, index=['Decision Tree'])
print(results_df)
```
